```
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```
im_1 = cv2.imread("1.png", cv2.IMREAD_GRAYSCALE)
im_2 = cv2.imread("2.png", cv2.IMREAD_GRAYSCALE)

im_h = cv2.cvtColor(im_1, cv2.COLOR_BGR2RGB)
im_ch = cv2.cvtColor(im_2, cv2.COLOR_BGR2RGB)

h_h = cv2.calcHist([im_h], [0], None, [256], [0, 256])
c_h = cv2.calcHist([im_ch], [0], None, [256], [0, 256])
```

```
plt.figure(figsize=(20,10))

plt.subplot(2,2,1)
plt.imshow(im_h, cmap = 'gray')

plt.subplot(2,2,2)
plt.plot(h_h); plt.xlim([0, 255]); plt.grid(); plt.plot();

plt.show()

plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
plt.imshow(im_ch,cmap = 'gray')

plt.subplot(2,2,2)
plt.plot(c_h); plt.xlim([0, 255]);  plt.grid(); plt.plot();


plt.show()
```

```
h_bl3 = cv2.blur(im_h, (3,3))
h_bl5 = cv2.blur(im_h, (5,5))
h_bl9 = cv2.blur(im_h, (9,9))
h_bl15 = cv2.blur(im_h, (15,15))

c_bl3 = cv2.blur(im_ch, (3,3))
c_bl5 = cv2.blur(im_ch, (5,5))
c_bl9 = cv2.blur(im_ch, (9,9))
c_bl15 = cv2.blur(im_ch, (15,15))
```

```
h_bl3_h = cv2.calcHist([h_bl3], [0], None, [256], [0, 256])
h_bl5_h = cv2.calcHist([h_bl5], [0], None, [256], [0, 256])
h_bl9_h = cv2.calcHist([h_bl9], [0], None, [256], [0, 256])
h_bl15_h = cv2.calcHist([h_bl15], [0], None, [256], [0, 256])

c_bl3_h = cv2.calcHist([c_bl3], [0], None, [256], [0, 256])
c_bl5_h = cv2.calcHist([c_bl5], [0], None, [256], [0, 256])
c_bl9_h = cv2.calcHist([c_bl9], [0], None, [256], [0, 256])
c_bl15_h = cv2.calcHist([c_bl15], [0], None, [256], [0, 256])
```

```
plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
plt.imshow(h_bl3, cmap = 'gray')

plt.subplot(2,2,2)
plt.plot(h_bl3_h); plt.xlim([0, 255]); plt.grid(); plt.plot();
plt.show()

###################################################################
plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
plt.imshow(h_bl5, cmap = 'gray')

plt.subplot(2,2,2)
plt.plot(h_bl5_h); plt.xlim([0, 255]); plt.grid(); plt.plot();
plt.show()

###################################################################
plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
plt.imshow(h_bl9, cmap = 'gray')

plt.subplot(2,2,2)
plt.plot(h_bl9_h); plt.xlim([0, 255]); plt.grid(); plt.plot();
plt.show()

###################################################################
plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
plt.imshow(h_bl15, cmap = 'gray')

plt.subplot(2,2,2)
plt.plot(h_bl15_h); plt.xlim([0, 255]); plt.grid(); plt.plot();
plt.show()


###################################################################
###################################################################
plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
plt.imshow(c_bl3, cmap = 'gray')

plt.subplot(2,2,2)
plt.plot(c_bl3_h); plt.xlim([0, 255]); plt.grid(); plt.plot();
plt.show()


###################################################################
plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
plt.imshow(c_bl5, cmap = 'gray')

plt.subplot(2,2,2)
plt.plot(c_bl5_h); plt.xlim([0, 255]); plt.grid(); plt.plot();
plt.show()


###################################################################
plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
plt.imshow(c_bl9, cmap = 'gray')

plt.subplot(2,2,2)
plt.plot(c_bl9_h); plt.xlim([0, 255]); plt.grid(); plt.plot();
plt.show()


###################################################################
plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
plt.imshow(c_bl15, cmap = 'gray')

plt.subplot(2,2,2)
plt.plot(c_bl15_h); plt.xlim([0, 255]); plt.grid(); plt.plot();
plt.show()
```
