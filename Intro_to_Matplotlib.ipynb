```
import matplotlib.pyplot as plt

import numpy as np

import pandas as pd

import seaborn as sns

from matplotlib.axes._axes import _log as matplotlib_axes_logger
matplotlib_axes_logger.setLevel('ERROR')
```

```
t = t = np.arange(0.0, 2.0, 0.01)
```

```
s1 = np.sin(2*np.pi*t)
s2 = np.sin(4*np.pi*t)
s3 = np.sin(8*np.pi*t)
s4 = np.sin(16*np.pi*t)
```

```
plt.figure(figsize=(10,10))
ax1 = plt.subplot(221)
ax1.plot(t, s1, c='r',linewidth=4.0)
ax1.set(xlabel='Час (с)', ylabel='Напруга (mВ)', title='Сінус функція з частотою 2pi')
ax1.grid()

ax2 = plt.subplot(222)
ax2.plot(t, s2, c='c',linewidth=3.0)
ax2.set(xlabel='Час (с)', ylabel='Напруга (mВ)', title='Сінус функція з частотою 4pi')
ax2.grid()

ax3 = plt.subplot(223)
ax3.plot(t, s3, c='m',linewidth=2.0)
ax3.set(xlabel='Час (с)', ylabel='Напруга (mВ)', title='Сінус функція з частотою 8pi')
ax3.grid()

ax4 = plt.subplot(224)
ax4.plot(t, s4, c='k',linewidth=1.0)
ax4.set(xlabel='Час (с)', ylabel='Напруга (mВ)', title='Сінус функція з частотою 16pi')
ax4.grid()

plt.show()
```
