```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```
img=cv2.cvtColor(cv2.imread("1_8.png",cv2.IMREAD_GRAYSCALE),cv2.COLOR_BGR2RGB)
```

```
plt.figure(figsize=(10,10))
plt.imshow(img)
plt.show()
```

```
bin_img = img.copy()
threshold = 10
max_val = 255
min_val = 0

for i in range(0,img.shape[0],1):
    for j in range(0,img.shape[1],1):
      for k in range(0,3):
        if (img[i][j][k] > threshold):
            bin_img[i][j][k] = max_val
        else:
            bin_img[i][j][k] = min_val
plt.figure(figsize=(10,10))
plt.imshow(bin_img)
plt.show();
```

```
kernel = np.ones((2,2),np.uint8)
erosion = cv2.erode(bin_img,kernel)
plt.figure(figsize=(10,10))
plt.imshow(erosion, cmap='gray')
plt.show();
```

```
dil = cv2.dilate(bin_img,kernel)


plt.figure(figsize=(5,5))
plt.imshow(dil, cmap='gray')
plt.show();
```

```
im=cv2.cvtColor(cv2.imread("street.jpg",cv2.IMREAD_GRAYSCALE),cv2.COLOR_BGR2RGB)
```

```
plt.figure(figsize=(5,5))
plt.imshow(im, cmap='gray')
plt.show();
```

```
bin_im = im.copy()
thr = 170
max_val = 255
min_val = 0

for row in range(im.shape[0]):
    for col in range(im.shape[1]):
      for k in range(0,3):
        if (im[row][col][k]) > thr:
            bin_im[row][col][k] = max_val
        else:
            bin_im[row][col][k] = min_val

plt.figure(figsize=(5,5))
plt.imshow(bin_im)
plt.show()
```

```
kern= np.ones((2,2),np.uint8)
eros=cv2.erode(bin_im,kern)

plt.figure(figsize=(5,5))
plt.imshow(eros, cmap='gray')
plt.show()
```

```
openin= cv2.morphologyEx(eros,cv2.MORPH_OPEN,kern)

plt.figure(figsize=(8,8))
plt.imshow(openin, cmap='gray')
plt.show()
```
